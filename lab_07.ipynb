```julia
using Plots
using DifferentialEquations
theme(:wong)
```

```julia

N = 888;
x0 = 18;
```

```julia
g(t) = 0.81;
v(t) = 0.0003;
fun(x,p,t) = (g(t)+v(t)*x)*(N-x)
tspan = (0,10);
pr = ODEProblem(fun, x0, tspan);
sol = solve(pr, timeseries_steps = 0.1);

pl1 = plot(sol,
label = false)
```

```julia
savefig(pl1,"11.png")
```
